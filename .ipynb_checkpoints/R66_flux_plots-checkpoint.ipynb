# Choosing a Flux Point

In a previous notebook `R66_source_scanning.ipynb` I've laid out several simulations to try to understand the effective thermal neutron flux that is present in the UMN setup for Run 66, and modifications thereof.  Here I plot select versions of those fluxes and more carefully present the important points.

## Adding the Walls

It was important to add the walls.  It was found that the albedo does increase the flux for the locations closest to the wall by about a factor of 2 or so.  It is also seen below that changing the bath to pure $^3$He causes a dramatic (factor of 20 or so) decrease in flux.  This indicates that in the future, to get better agreement with measurements, it may be necessary to implement the mixing chamber and still in the fridge model in Geant4. 

```python
import scanData as sd
import numpy as np
import getPositionScan_HtoL as gp

#a function to get three rows of data x y z
data = sd.getScanData('data/Run66Sim_FluxScan0.txt')
dataF4 = sd.getScanData('data/Run66Sim_FluxScan4.txt')
dataF7 = sd.getScanData('data/Run66Sim_FluxScan7.txt')

justy = np.zeros(np.shape(data['xx'])) #get just the y-coordinates of the above data
for i,x in enumerate(data['xx']):
  justy[i] = gp.getPosition(i+1,10,False)[1]

#also get the experimental measurements of flux
import dataPython as dp
measF = dp.getXYdata_wYerr('data/Run66_Flux_measurements.txt')
meas_justy = np.zeros(np.shape(measF['xx']))
meas_justy[0] = gp.getPosition(1,10,False)[1]
meas_justy[1] = gp.getPosition(10,10,False)[1]
```
